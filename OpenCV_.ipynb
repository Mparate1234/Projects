# Blur

```python

img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2HSV)  # Corrected here
figure_size = 50
result = cv2.blur(img3, (figure_size,figure_size))  # Blur function requires a tuple for kernel size

plt.figure(figsize=(11,6))
plt.subplot(121), plt.imshow(cv2.cvtColor(img3, cv2.COLOR_HSV2RGB)), plt.title("Original")  # Fixed the cv2.cvtColor function here
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(cv2.cvtColor(result, cv2.COLOR_HSV2RGB)), plt.title("Blurred")  # Also fixed the cv2.cvtColor function here
plt.xticks([]), plt.yticks([])
plt.show()
```

# Rotate

```python
height, width = img3.shape[:2]
r = cv2.getRotationMatrix2D((width/2, height/2), 30, 1)  # 30 is the rotation angle, 1 is the scale
rotate = cv2.warpAffine(img3, r, (width, height))
plt.imshow(rotate)
plt.show()
```
